```python
import pandas as pd
df = pd.read_csv("Cars93.csv")
df.tail()
```

```python
def handle_missing_values_alternative(df):
    # Create a copy of the dataframe
    df_copy = df.copy()

    for col in df_copy.select_dtypes(include=['float64', 'int64']).columns:
        df_copy[col] = df_copy[col].fillna(df_copy[col].median())

    for col in df_copy.select_dtypes(include=['object']).columns:
        df_copy[col] = df_copy[col].fillna('Unknown')

    return df_copy

# Usage
df = handle_missing_values_alternative(df)

df.tail()
```

```python
 # Rounding numeric columns to 2 decimal places

def reduce_noise(df):
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        df[col] = df[col].round(2)
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].str.strip().str.lower()

    return df

df= reduce_noise(df)
df.tail()
```

```python
from sklearn.preprocessing import LabelEncoder
# Convert Categorial value to Numerical value
def encode_categorical_features(df):
    label_encoders = {}
    for col in df.select_dtypes(include=['object']).columns:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le
    return df, label_encoders

encoded_data, encoders = encode_categorical_features(df)

df.tail()
```

```python
from sklearn.model_selection import train_test_split

def split_data(df, train_ratio=0.7, validation_ratio=0.2, test_ratio=0.1):
    train_data, temp_data = train_test_split(df, test_size=(1 - train_ratio), random_state=42)
    validation_data, test_data = train_test_split(temp_data, test_size=(test_ratio / (test_ratio + validation_ratio)), random_state=42)
    return train_data, validation_data, test_data


train_data, validation_data, test_data = split_data(df)

train_data
```
